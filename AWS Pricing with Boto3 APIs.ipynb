```python
import boto3
import json
import openpyxl
import logging

# Create a logger
logger = logging.getLogger(__name__)
logger.setLevel(logging.DEBUG)

# Create a formatter with date and time information
formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')

# Create a file handler and set its formatter
file_handler = logging.FileHandler('aws_region_pricing.log')
file_handler.setFormatter(formatter)

# Add the file handler to the logger
logger.addHandler(file_handler)


def get_elasticache_prices(region, engine):
    client = boto3.client('pricing')

    response = client.get_products(
        ServiceCode='AmazonElastiCache',
        Filters=[
            {'Type': 'TERM_MATCH', 'Field': 'location', 'Value': region},
            {'Type': 'TERM_MATCH', 'Field': 'cacheEngine', 'Value': engine},
        ],
        FormatVersion='aws_v1'
    )

    prices = []
    for price in response['PriceList']:
        parsed = json.loads(price)
        instance_type = parsed['product']['attributes']['instanceType']
        region_code = parsed['product']['attributes'].get('location')
        terms = parsed['terms']

        ondemand_price = None
        reserved_price = None

        if 'OnDemand' in terms:
            ondemand_priceDimensions = list(terms['OnDemand'].values())[0].get('priceDimensions')
            if ondemand_priceDimensions:
                ondemand_priceDescription = ondemand_priceDimensions[list(ondemand_priceDimensions.keys())[0]].get('description')
                ondemand_price = ondemand_priceDimensions[list(ondemand_priceDimensions.keys())[0]]\
                    .get('pricePerUnit', {}).get('USD')

        if 'Reserved' in terms:
            reserved_priceDimensions = terms['Reserved'].values()
            for dimension in reserved_priceDimensions:
                dimension_priceDimensions = dimension.get('priceDimensions')
                lease_contract_length = dimension.get('termAttributes', {}).get('LeaseContractLength')
                purchase_options = dimension.get('termAttributes', {}).get('PurchaseOption')
                if dimension_priceDimensions:
                    for key in dimension_priceDimensions:
                        reserved_priceDescription = dimension_priceDimensions[key].get('description')
                        reserved_priceUnit = dimension_priceDimensions[key].get('unit')#Here
                        reserved_price = dimension_priceDimensions[key].get('pricePerUnit', {}).get('USD')
                        if (reserved_priceUnit == 'Quantity') and (purchase_options in ('All Upfront', 'Heavy Utilization')) and (lease_contract_length == '3yr'): #Here
                            prices.append((region_code, instance_type, ondemand_price, reserved_price,lease_contract_length,purchase_options ))

        # Log the instance type and prices
#         logger.info(f"Instance Type: {instance_type}, Region: {region_code}, On-Demand Price: {ondemand_price}, Reserved Price: {reserved_price}")

    return prices

# Create a new workbook
workbook = openpyxl.Workbook()

# Define the list of regions to query
# regions = ['US East (N. Virginia)', 'US West (Oregon)', 'EU (Ireland)']
regions = ['US East (Ohio)', 
           'US East (N. Virginia)', 
           'US West (N. California)', 
           'US West (Oregon)',
           'Africa (Cape Town)',
           'Asia Pacific (Hong Kong)',
           'Asia Pacific (Hyderabad)',
           'Asia Pacific (Jakarta)',
           'Asia Pacific (Melbourne)',
           'Asia Pacific (Mumbai)',
           'Asia Pacific (Osaka)',
           'Asia Pacific (Seoul)', 
           'Asia Pacific (Singapore)', 
           'Asia Pacific (Sydney)', 
           'Asia Pacific (Tokyo)', 
           'Canada (Central)', 
           'China (Beijing)', 
           'China (Ningxia)',                                                        
           'EU (Frankfurt)', 
           'EU (Ireland)', 
           'EU (London)', 
           'EU (Milan)', 
           'EU (Paris)', 
           'EU (Stockholm)', 
           'Europe (Zurich)',
           'Middle East (Bahrain)',
           'Middle East (UAE)',
           'AWS GovCloud (US-East)', 
           'AWS GovCloud (US-West)']


for region in regions:
    # Add headers to the worksheet
    worksheet = workbook.create_sheet(title=region+" B")
    worksheet.append(['Region', 
                      'Type', 
                      'OnDemand Price', 
                      'Reserved Price',
                      'Lease Contract Length',
                      'Purchase Options'])

    # Call the function and write the results to the worksheet
    prices = get_elasticache_prices(region, 'Redis')
    logger.info(f"Extrating Data For: {region} Please wait!")
    for price in prices:
        worksheet.append(price)
    
        

# Save the workbook with the region name appended to the file name
workbook.save(f"All_Region_Output.xlsx")

# Log that the file was saved
logger.info(f"Output file saved for region {region}")
```
